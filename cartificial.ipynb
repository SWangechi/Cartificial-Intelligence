## 3. Model Training

### 3.1 Collaborative Filtering

Using the surprise library to create a collaborative filtering recommendation system based on user-product interaction

```python
from surprise import SVD, Dataset, Reader, accuracy
from surprise.model_selection import cross_validate, train_test_split

# Prepare data for collaborative filtering
reader = Reader(rating_scale=(0,5))
data = Dataset.load_from_df(df[['CustomerID', 'ProductID', 'ProductRating']], reader)
```

### 3.2 Build and evaluate the collaborative filtering model

```python
svd_model = SVD()
cross_validate(svd_model, data, measures=['RMSE', 'MAE'], cv=5, verbose=True)
```

#### 3.3 Split the data into train and test sets

```python
trainset, testset = train_test_split(data, test_size=0.2)
```

#### 3.4 Train the model on the entire dataset

```python
# trainset = data.build_full_trainset()
svd_model.fit(trainset)
```

#### 3.5 Predict ratings for the test set

```python
predictions = svd_model.test(testset)
```

#### 3.6 Evaluate the accuracy of the model

```python
rmse = accuracy.rmse(predictions)
mae = accuracy.mae(predictions)
```

#### 3.7 Predict function for recommendation

```python
def recommend_products(user_id, model, data, top_n=5):
    product_ids = df['ProductID'].unique()
    predictions = [model.predict(user_id, pid).est for pid in product_ids]
    recommend_products = sorted(list(zip(product_ids, predictions)), key=lambda x: x[1], reverse=True)[:top_n]
    return recommend_products

# Recommend products for user 9691
recommend_products(9691, svd_model, data)
```

**Evaluating Accuracy:**

- accuracy.rmse(predictions) calculates the Root Mean Squared Error.

## 4. Content-Based Filtering

### 4.1 Vectorize Product Description

Using TF-IDF vectorizer to recommend products based on their descriptions or other textual data.

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

df['ProductDesc'] = 'Product ' + df['ProductID'].astype(str) 

tfidf = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf.fit_transform(df['ProductDesc'])

# Compute cosine similarity
cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)
```

### 4.2 Predict function for recommendation

```python
def get_content_based_recommendations(product_id, cosine_sim=cosine_sim, top_n=5):
    idx=df[df['ProductID']== product_id].index[0]
    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)[1:top_n+1]
    product_indices = [i[0] for i in sim_scores]
    return df.iloc[product_indices][['ProductID', 'Total']]

# Recommend products similar to ProductID 32
get_content_based_recommendations(32)
```

## 5. Hybrid Approach

Combining collaborative filtering and content-based filtering for better accuracy.

Using weigh scores to recommend products

```python
def hybrid_recommendation(user_id, product_id, model, cosine_sim, alpha=0.5, top_n=5):
    content_recs = get_content_based_recommendations(product_id, cosine_sim, top_n)
    collab_recs = recommend_products(user_id, model, data, top_n)
    
    combined = pd.DataFrame({
        'ProductID': [rec[0] for rec in collab_recs],
        'Collab_Score': [rec[1] for rec in collab_recs]
    })
    
    content_scores = content_recs.set_index('ProductID')
    combined = combined.set_index('ProductID').join(content_scores, how='inner').reset_index()
    combined['Final_Score'] = alpha * combined['Collab_Score'] + (1 - alpha) * combined['Total']
    return combined.sort_values(by='Final_Score', ascending=False).head(top_n)

# Example: Hybrid recommendation
hybrid_recommendation(9691, 32, svd_model, cosine_sim)
```

```python
def hybrid_recommendation(user_id, product_id, model, cosine_sim, alpha=0.5, top_n=5):
    # Get content-based recommendations
    content_recs = get_content_based_recommendations(product_id, cosine_sim, top_n)
    if content_recs.empty:
        raise ValueError("Content-based recommendations returned an empty result.")

    # Get collaborative recommendations
    collab_recs = recommend_products(user_id, model, data, top_n)
    if not collab_recs:
        raise ValueError("Collaborative recommendations returned an empty result.")

    # Create DataFrame for collaborative recommendations
    collab_df = pd.DataFrame({
        'ProductID': [str(rec[0]) for rec in collab_recs],  # Convert to string
        'Collab_Score': [rec[1] for rec in collab_recs]
    })

    # Ensure consistent data types for ProductID
    content_recs['ProductID'] = content_recs['ProductID'].astype(str)

    # Merge content and collaborative scores
    combined = pd.merge(collab_df, content_recs[['ProductID', 'Total']], on='ProductID', how='inner')

    # Debugging: Check data at each step
    print("\n--- Debugging Information ---")
    print("Collaborative Recommendations:\n", collab_df)
    print("Content-Based Recommendations:\n", content_recs[['ProductID', 'Total']])
    print("Combined DataFrame After Merge:\n", combined)
    print("-----------------------------\n")

    # Handle case where no overlap exists
    if combined.empty:
        print("No overlapping products found. Returning top collaborative recommendations.")
        collab_df['Final_Score'] = collab_df['Collab_Score']  # Use collaborative scores as fallback
        return collab_df.sort_values(by='Collab_Score', ascending=False).head(top_n)

    # Normalize scores
    combined['Collab_Score'] = combined['Collab_Score'] / combined['Collab_Score'].max()
    combined['Total'] = combined['Total'] / combined['Total'].max()

    # Calculate final scores
    combined['Final_Score'] = alpha * combined['Collab_Score'] + (1 - alpha) * combined['Total']

    # Sort and return top recommendations
    return combined.sort_values(by='Final_Score', ascending=False).head(top_n)

# Example call
try:
    result = hybrid_recommendation(9691, 32, svd_model, cosine_sim)
    print(result)
except ValueError as e:
    print("Error:", e)
```
